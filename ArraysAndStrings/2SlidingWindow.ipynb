>Whenever a problem description talks about subarrays, you should figure out if sliding window is a good option.

1. Find the longest subarray with a sum less than or equal to k
2. Find the longest substring that has at most one "0"
3. Find the number of subarrays that have a product less than k



### Example 1: Given an array of positive integers nums and an integer k, find the length of the longest subarray whose sum is less than or equal to k. 

```python
class Solution():
    def lenSub(self, nums, k):
        # curr tracks the sum of the current window and ans is the len of the subarray
        left = curr = ans = 0
        for right in range(len(nums)):
            curr += nums[right]
            while curr > k:
                curr -= nums[left]
                left += 1
            ans = max(ans, right - left + 1)
        
        return ans 
```

```python
sol = Solution()
result = sol.lenSub([3, 1, 2, 7, 4, 2, 1, 1, 5], 8)
print(result)
```
